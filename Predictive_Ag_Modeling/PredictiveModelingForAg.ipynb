# Sowing Success: How Machine Learning Helps Farmers Select the Best Crops

![Farmer in a field](./graphics/farmer_in_a_field.jpg)

Measuring essential soil metrics such as nitrogen, phosphorous, potassium levels, and pH value is an important aspect of assessing soil condition. However, it can be an expensive and time-consuming process, which can cause farmers to prioritize which metrics to measure based on their budget constraints.

Farmers have various options when it comes to deciding which crop to plant each season. Their primary objective is to maximize the yield of their crops, taking into account different factors. One crucial factor that affects crop growth is the condition of the soil in the field, which can be assessed by measuring basic elements such as nitrogen and potassium levels. Each crop has an ideal soil condition that ensures optimal growth and maximum yield.

A farmer reached out to you as a machine learning expert for assistance in selecting the best crop for his field. They've provided you with a dataset called `soil_measures.csv`, which contains:

- `"N"`: Nitrogen content ratio in the soil
- `"P"`: Phosphorous content ratio in the soil
- `"K"`: Potassium content ratio in the soil
- `"pH"` value of the soil
- `"crop"`: categorical values that contain various crops (target variable).

Each row in this dataset represents various measures of the soil in a particular field. Based on these measurements, the crop specified in the `"crop"` column is the optimal choice for that field.  

In this project, you will build multi-class classification models to predict the type of `"crop"` and identify the single most importance feature for predictive performance.

```python
# All required libraries are imported here for you.
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics

# Load the dataset
crops = pd.read_csv("./data/soil_measures.csv")

print(crops.shape)
crops.head()
```

## Visualize class distributions

How does each predictor distribute itself within each crop? How do these distributions look relative to each other?

Intuitively, we'd expect the predictor that shows the most variation over the different crop types to be the most predictive.

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_crop_pred(df, pred, plot_title=""):
    """ Creates a boxplot of the distributions of a predictor (pred) across
    types of crop

    Args:
      df (dataframe): has 'crop' and 'pred' columns corresponding to crop type and
        predictor respectively
      pred (str): predictor being evaluated at each crop type
      plot_title (str): plot title

    Returns:
      matplotlib.axes._axes.Axes: axes plot object for the boxplot
    """
    # https://www.statology.org/seaborn-boxplot-order-x-axis/
    group_medians = df.groupby(['crop'], observed=False)[pred].median().sort_values(ascending=True)
    
    pred_plot = sns.boxplot(data=df, x='crop', y=pred, hue="crop", order=group_medians.index)
    pred_plot.axhline(y = df[pred].median(), color='red')  # add grand median reference line
    pred_plot.set_title(plot_title)
    plt.xticks(rotation=90)
    return pred_plot

# nitrogen
nitrogen_plot = plot_crop_pred(crops, 'N', "Nitrogen distribution by crop type \n(Red line is grand median)")
```

```python
# phosphorous
phosphorous_plot = plot_crop_pred(crops, 'P', "Phosphorous distribution by crop type \n(Red line is grand median)")
```

```python
# potassium
potassium_plot = plot_crop_pred(crops, 'K', "Potassium distribution by crop type \n(Red line is grand median)")
```

```python
# pH
ph_plot = plot_crop_pred(crops, 'ph', "pH distribution by crop type \n(Red line is grand median)")
```

## Encode the target column

The `crop` column is categorical and needs to be encoded.

```python
crops['crop'] = crops['crop'].astype("category")
crops['crop_code'] = crops['crop'].cat.codes
print(crops.head(), '\n')
print(crops.tail(), '\n')
```

## Strategy

### Metric

The *negative log likelihood* or *cross-entropy* is used as the loss function for logistic regression in scikit, so this would be the most natural choice for the metric we use to evaluate model performance, but **accuracy** is more intuitive, so it will be used here.

### Workflow

1. Build a set of single parameter logistic regression models
2. Evaluate each set of models using 5-Fold CV ran 3 times using accuracy as the metric
3. Create a boxplot with each predictor on the x-axis and accuracy on the y-axis

The boxplot will allow us to see how each of the single-predictor models relative to each other. From this plot, the best model will be selected as the parameter with the most importance feature for predictive performance.

```python
rand_state = 711  # for reproducibility
# predictor and target vectors
X = crops.drop(["crop", "crop_code"], axis="columns")
y = crops['crop_code']
# split to train and test sets - do BEFORE scaling
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=rand_state)
X.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler

cv_folds = 5         # as specified above
repeat_cv_folds = 3  # as specified above

# last 2 columns are the response and encoded response
predictors = list(crops.columns)[:-2]

accuracies_cv_all = []
accuracies_cv_preds = []
accuracy_cv_means = []
# cross_ents = []

# build logistic regression models from each single predictor
for pred in predictors:
    # model = LogisticRegression(solver='lbfgs', max_iter=3000)
    model = LogisticRegression()
    scaler = StandardScaler()
    # scale the train and test predictor vectors
    X_train_scaled = scaler.fit_transform(X_train[pred].to_numpy().reshape(-1, 1))
    X_test_scaled = scaler.fit_transform(X_test[pred].to_numpy().reshape(-1, 1))
    
    X_train_pred = X_train_scaled
    model.fit(X_train_pred, y_train)
    # define model eval method, 5-fold cv x 3
    cv = RepeatedStratifiedKFold(n_splits=cv_folds, n_repeats=3, random_state=rand_state)
    # evaluate model
    scores_acc = cross_val_score(model, X_train_pred, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
    # scores_cross_ent = cross_val_score(model, X_train_pred, y_train, scoring='neg_log_loss', cv=cv, n_jobs=-1)
    cv_score_acc = np.mean(scores_acc)
    # cv_score_cross_ent = np.mean(scores_cross_ent)
    accuracies_cv_all.extend(scores_acc)
    accuracies_cv_preds.extend([pred] * (cv_folds * repeat_cv_folds))
    accuracy_cv_means.append(cv_score_acc)
    # cross_ents.append(cv_score_cross_ent)
    print(f"For the {pred} predictor, mean CV model accuracy = {round(cv_score_acc, 5)}, random accuracy = {round(1/len(crop_counts), 5)}")
    # print(f"For the {pred} predictor, mean CV model cross ent = {round(cv_score_cross_ent, 5)}, random accuracy = {round(1/len(crop_counts), 5)} \n")

print('**************************************************')

# evaluate best predictor
best_accuracy = max(accuracy_cv_means)
best_acc_index = accuracy_cv_means.index(best_accuracy)
best_acc_pred = predictors[best_acc_index]
# best_cross_ent = max(cross_ents)
# best_cross_end_index = cross_ents.index(best_cross_ent)
# best_cross_ent_pred = predictors[best_cross_end_index]

print(f"best Logistic Regression predictor by accuracy is {best_acc_pred} with accuracy of {best_accuracy}")
# print(f"best predictor by cross entropy is {best_cross_ent_pred} with cross entropy of {best_cross_ent}")
```

```python
logistic_reg_final_plot = sns.boxplot(data=df_acc_by_pred, x='predictor', y='accuracy', hue='predictor')
# print(type(logistic_reg_final_plot))  # <class 'matplotlib.axes._axes.Axes'>
logistic_reg_final_plot.set_title("Logistic Regression CV Accuracies by Predictor")
plt.show()
```

## Verify with LDA

When choosing between **logistic regression** and **linear discrimant analysis (LDA)** when the number of target classes > 2, the later is often the preferred model as described here:

https://stats.stackexchange.com/questions/188416/discriminant-analysis-vs-logistic-regression#188420

Having run the logistic regressions, the same analysis is repeated below in order to verify the initial results.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# reset the list holding the results
accuracies_cv_all = []
accuracies_cv_preds = []
accuracy_cv_means = []

# build LDA models from each single predictor
for pred in predictors:
    model = LDA()
    scaler = StandardScaler()
    # scale the train and test predictor vectors
    X_train_scaled = scaler.fit_transform(X_train[pred].to_numpy().reshape(-1, 1))
    X_test_scaled = scaler.fit_transform(X_test[pred].to_numpy().reshape(-1, 1))
    
    X_train_pred = X_train_scaled
    model.fit(X_train_pred, y_train)
    # define model eval method, 5-fold cv x 3
    cv = RepeatedStratifiedKFold(n_splits=cv_folds, n_repeats=3, random_state=rand_state)
    # evaluate model
    scores_acc = cross_val_score(model, X_train_pred, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
    # scores_cross_ent = cross_val_score(model, X_train_pred, y_train, scoring='neg_log_loss', cv=cv, n_jobs=-1)
    cv_score_acc = np.mean(scores_acc)
    # cv_score_cross_ent = np.mean(scores_cross_ent)
    accuracies_cv_all.extend(scores_acc)
    accuracies_cv_preds.extend([pred] * (cv_folds * repeat_cv_folds))
    accuracy_cv_means.append(cv_score_acc)
    # cross_ents.append(cv_score_cross_ent)
    print(f"For the {pred} predictor, mean CV model accuracy = {round(cv_score_acc, 5)}, random accuracy = {round(1/len(crop_counts), 5)}")

print('**************************************************')

# evaluate best predictor
best_accuracy = max(accuracy_cv_means)
best_acc_index = accuracy_cv_means.index(best_accuracy)
best_acc_pred = predictors[best_acc_index]

print(f"best LDA predictor by accuracy is {best_acc_pred} with accuracy of {best_accuracy}")
```

```python
df_lda_acc_by_pred = pd.DataFrame({'predictor': accuracies_cv_preds,
                                   'accuracy': accuracies_cv_all})
df_lda_acc_by_pred.head(17)
```

```python
lda_reg_final_plot = sns.boxplot(data=df_lda_acc_by_pred, x='predictor', y='accuracy', hue='predictor')
lda_reg_final_plot.set_title("LDA CV Accuracies by Predictor")
plt.show()
```

Looks like LDA does a little better than logistic regression, but they both identify potassium (K) as the best single predictor of optimal crop type.
